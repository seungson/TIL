```python
import numpy as np
import pandas as pd 

import matplotlib.pyplot as plt
import seaborn as sns

from scipy import stats

plt.style.use(['seaborn'])
```

```python
def show_sample_dist(N, size):       
    mean_list = []
    for i in range(1,10):
        plt.subplot(int(f'33{i}'))
        data = np.sqrt(np.random.randn(size)**2)
        mean_list.append(data.mean())
        sns.histplot(data, alpha=0.55)
        plt.axvline(np.mean(data), linestyle='--', c='r', alpha=0.55, linewidth=0.7)
        plt.text(x=np.mean(data)+0.02, y=0, s=f'{np.mean(data):.2f}', c='r', size=10)
        plt.ylabel('')
    plt.suptitle(f'<Distribution of 9 samples(size={size}>')
    plt.tight_layout()
    plt.show()
    
    for i in range(10,N):
        data = np.sqrt(np.random.randn(size)**2)
        mean_list.append(data.mean())
    
    plt.figure(figsize=(9,4))
    
    plt.subplot(121)
    M = np.sqrt(np.random.randn(1000000)**2)
    sns.histplot(M)
    plt.axvline(np.mean(M), linestyle='--', c='r', alpha=0.66, linewidth=0.7)
    plt.text(x=np.mean(M), y=0, s=f'{np.mean(M):.2f}', c='r', size=12)
    plt.title(f'<Distribution of population>')
    plt.ylabel('')
    
    plt.subplot(122)
    sns.histplot(mean_list)
    plt.title(f'<Distribution of the sample mean(size={size})>')
    plt.axvline(np.mean(mean_list), linestyle='--', c='r', alpha=0.66, linewidth=0.7)
    std = np.std(mean_list)
    mean = np.mean(mean_list)
    plt.axvspan(xmin=mean-std, xmax=mean+std, color='lightcoral', alpha=0.45)
    plt.axvspan(xmin=mean-std*2, xmax=mean+std*2, color='mistyrose', alpha=0.45)
    plt.text(x=np.mean(mean_list), y=0, s=f'{np.mean(mean_list):.2f}', c='r', size=12)
    plt.ylabel('')
    
    plt.show()
    print(f'Mean of sample mean : {mean:.2f}')
    print(f'Standard deviation of sample mean : {std:.2f}')
    
```

```python
show_sample_dist(N=10000, size=1)
```
